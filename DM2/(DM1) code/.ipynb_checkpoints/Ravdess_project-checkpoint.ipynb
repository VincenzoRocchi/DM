```python
sns.barplot(x = percent_nan.index, y= percent_nan)
plt.show()
```

```python
#checking the categorical attributes and their unique values fto discretize them

print(pd.unique(df["emotion"])) #too many unique values for the emotion column, solving it later
print(pd.unique(df["vocal_channel"]))
print(pd.unique(df["emotional_intensity"]))
print(pd.unique(df["repetition"]))
print(pd.unique(df["statement"]))
print(pd.unique(df["sex"]))
```

```python
#defining the maps to convert the categorical variables in numerical ones

vocal_map = {"speech": 1, "song": 2}
emotional_int_map = {"normal": 1, "strong" :2}
repetition_map = {"1st": 1, "2nd" :2}
sex_map = {"F": 1, "M" :2}
statement_map = {"Dogs are sitting by the door" :1, "Kids are talking by the door":2}
emotion_map = {'fearful':0, 'angry':1, 'happy':2, 'surprised':3, 'neutral':4, 'calm':5, 'sad':6, 'disgust':7}
```

```python
#--------------
#using the dummies method to convert the emotion attribute

#emotion_dummies = pd.get_dummies(df.emotion)
#df_copy = pd.concat([df, emotion_dummies], axis = "columns")

#df_copy.drop(["emotion"], axis = 1, inplace = True)
#----------------
```

```python
#replacing the maps and creating the corr graph

df_copy.replace({"vocal_channel": vocal_map,
            "emotional_intensity": emotional_int_map,
            "repetition": repetition_map,
            "sex": sex_map,
            "statement": statement_map,
            "emotion" : emotion_map}, inplace = True)

df_copy.corr(method="pearson").style.background_gradient(cmap='coolwarm',
                                                               vmin=-1, vmax=1, axis=None,).format(precision=2)
```
